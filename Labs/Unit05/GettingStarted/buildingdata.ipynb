# Building Datasets

## 1. Import Libraries

```python
import pandas as pd
```

## 2. Load the Dataset

```python
df = pd.read_csv("data/NY-House-Dataset.csv")
df.head()
```

## 3. Modifying Columns

### Dropping Columns

```python
df = df.drop('BROKERTITLE', axis=1)
df.columns
```

```python
df = df[['TYPE', 'PRICE', 'BEDS', 'BATH', 'PROPERTYSQFT', 'ADDRESS', 'STATE',
       'MAIN_ADDRESS', 'LOCALITY', 'SUBLOCALITY']]
df.shape
```

### Adding Columns

```python
df1 = df.assign(PRICE_PER_SQFT = df['PRICE'] / df['PROPERTYSQFT'])

df1.head()
```

```python
df['PRICE_PER_SQFT'] = df['PRICE'] / df['PROPERTYSQFT']

df.head()
```

## 4. Sorting

```python
df.head()

df.sort_values('PRICE_PER_SQFT', ascending=False)
```

## 5. Removing Rows

```python
df['TYPE'].value_counts()
```

```python
df['TYPE'].unique()
```

```python
keep =['Condo for sale', 'House for sale', 'Townhouse for sale',
       'Co-op for sale', 'Multi-family home for sale', 'Mobile house for sale']
df = df.loc[df['TYPE'].isin(keep)]
df
```

```python
df['TYPE'].value_counts()
```

## 6. Map

```python
def remove_for_sale(word):
    index = word.find(" for sale")
    if index != -1:
        return word[:index]
    return word

df['TYPE'] = df['TYPE'].map(remove_for_sale)
df['TYPE'].value_counts()
```

## 7. Groupby

```python
df.groupby('TYPE')[["PRICE_PER_SQFT"]].mean()
```

```python
df.groupby(["TYPE", "SUBLOCALITY"])[["PRICE_PER_SQFT"]].mean()
```

## 8. Homework

```python
df = pd.read_csv("data/NY-House-Dataset.csv")
```

```python
def remove_brokered_by(word):
    s = "Brokered by "
    index = word.find(s)
    if index != -1:
        return word[:index] + word[index + len(s):]
    return word

dftemp = df.copy()

dftemp['BROKERTITLE'] = dftemp['BROKERTITLE'].map(remove_brokered_by)
dftemp['BROKERTITLE'].value_counts()
```

```python
dftemp = df.copy()

dftemp = dftemp.loc[dftemp['LOCALITY'].str.contains('County')]
dftemp.groupby('LOCALITY')[['PROPERTYSQFT']].max()
```

```python
dftemp = df.copy()

dftemp['BED_TO_BATH'] = dftemp['BEDS'] / dftemp['BATH']
dftemp.groupby('SUBLOCALITY')[['BED_TO_BATH']].mean().sort_values('BED_TO_BATH', ascending=False)
```
