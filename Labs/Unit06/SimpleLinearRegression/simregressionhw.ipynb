The provided data from the advertising dataset describes the number of units of a product sold given the various TV, radio, and newspaper advertising budgets (in thousands).

## 1. Library Imports and Data Loading

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv("data/advertising.csv")
```

## 2. Understanding the Data

```python
# TODO: Determine the features and label of the dataset
df.info()
```

So "sales" is the dependent variable.

## 3. Model 1

```python
# TODO: Create and train your model

X_train = df[["TV"]]
y_train = df["sales"]

model = LinearRegression()

model = model.fit(X_train, y_train)
```

```python
# TODO: Make the 2 predictions asked of you

model.predict([[100], [500]])
```

```python
# TODO: Plot your data and line of best fit
# Don't forget to include a graph title with plt.title(...)
plt.title("Sales vs TV Advertisements")
sns.scatterplot(data=df, x="TV", y="sales")
plt.plot(X_train, model.predict(X_train), color = "r")
```

```python
df.loc[df["sales"] == df["sales"].min()]

#No outliers to remove
```

```python
# Calculate the residuals
y_pred = model.predict(X_train)
residuals = y_train - y_pred
# Graph the residuals
# residuals = [r for r in residuals if r > -250 and r < 250]
sns.histplot(residuals)
```

```python
plt.title("Residuals vs. TV")
sns.scatterplot(data=X_train, x="TV", y=residuals)
```

The model seems to get worse as the TV advertisement gets bigger: the residuals are heteroscedastic.

```python
model.score(X_train, y_train)
```

This model is fairly accurate.
